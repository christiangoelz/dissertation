```python
import glob
import pickle 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# Define custom colormap colors
colors = [((255/255), (255/255), (255/255)),
          ((3/255), (199/255), (199/255))]
cmap_old  = mcolors.LinearSegmentedColormap.from_list('cmap_old', colors)

colors = [((255/255), (255/255), (255/255)),
          ((199/255), (3/255), (199/255))]
cmap_young  = mcolors.LinearSegmentedColormap.from_list('cmap_young', colors)

part = [] 
confmat = [] 
wd = '/media/chris/SEAGATE_BAC1/Science/Dissertation/Project2_DMDMotorControl/Results/EEG/Classification/Results_raw/EXCLUDE_Z/2CSP/LDA_puplished'
files = glob.glob(wd + '/na*Classification_All.pkl')
l = len(wd) + 1
ll = l + 4
for file in files: 
    with open(file, 'rb') as handle:
        participant = file[l:ll]
        part.append(file[l:ll])
        data = pickle.load(handle)
        confmat.append(np.mean(data['All'].conf_mat, axis = 0))
        
m = np.mean(confmat,axis = 0)

```

```python
fig, ax = plt.subplots(tight_layout=True,figsize=(4,3))
_ = sns.heatmap(m, annot=True, vmin = 0, vmax =1, yticklabels=False, xticklabels=False, cmap =cmap_old, ax=ax)
plt.show()
fig.savefig('/home/chris/Desktop/dissertation/figures/drafts/paper1_old.svg')
```
